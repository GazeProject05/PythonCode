```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.stats import multivariate_normal
pd.set_option('display.max_columns', None)

mypathA = 'Datasets/GazeA/'
mypathB = 'Datasets/GazeB/'
mydtaframesA = {}
mydtaframesB = {}


for index,files in enumerate(os.listdir(mypathA)):
    mydtaframesA["df_{}".format(index)] = pd.read_table('{}/{}'.format(mypathA, files),low_memory=False)

for index,files in enumerate(os.listdir(mypathB)):
    mydtaframesB["df_{}".format(index)] = pd.read_table('{}/{}'.format(mypathB, files),low_memory=False)
    
dfA = {}
dfB = {}

for index,frames in enumerate(mydtaframesA):
    dfA["{}".format(index)] = mydtaframesA[frames]
for frames in dfA:
    dfA[frames].fillna(method='ffill', inplace= True)
    

for index,frames in enumerate(mydtaframesB):
    dfB["{}".format(index)] = mydtaframesB[frames]
for frames in dfB:
    dfB[frames].fillna(method='ffill', inplace= True)
```
