```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy import stats
import sklearn
ls = np.linspace(0, 60, 1000)

#distribution = mlab.normpdf(ls, 0, 5) + mlab.normpdf(ls, 20, 10)
#distribution = (distribution * 1000).astype(int)
#distribution = distribution/distribution.sum()

#plt.plot(ls, distribution)

import seaborn as sns
sns.set_style("white")
%matplotlib inline
import pandas as pd
#show the distribution of the data as a histogram
#read our dataset
df = pd.read_excel("reading_all.xlsx")
#show first 5 examples (in BTC)
df.head(n=5)
data = df['PupilLeft']
data = data.str.replace(',','.').astype(float)
#data = df.pupilleft
#plot histogram
sns.distplot(data, bins=100, kde=True)
#try to fit a normal distribution to this data
sns.distplot(data, fit=stats.norm, bins=20, kde=False,)

#sklearn.mixture.GaussianMixture()
Min_graph = min(data)
Max_graph = max(data)
x = np.linspace(Min_graph, Max_graph, 2000)

from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components = 3, tol=0.000001)
gmm.fit(np.expand_dims(data, 1)) # Parameters: array-like, shape (n_samples, n_features), 1 dimension dataset so 1 feature
Gaussian_nr = 1
for mu, sd, p in zip(gmm.means_.flatten(), np.sqrt(gmm.covariances_.flatten()), gmm.weights_):
    print('Gaussian {:}: μ = {:.2}, σ = {:.2}, weight = {:.2}'.format(Gaussian_nr, mu, sd, p))
    g_s = stats.norm(mu, sd).pdf(x) * p
    plt.plot(x, g_s, label='gaussian sklearn');
    Gaussian_nr += 1
sns.distplot(data, bins=20, kde=False, norm_hist=True)
gmm_sum = np.exp([gmm.score_samples(e.reshape(-1, 1)) for e in x]) #gmm gives log probability, hence the exp() function
plt.plot(x, gmm_sum, label='gaussian mixture');
plt.legend();
```
